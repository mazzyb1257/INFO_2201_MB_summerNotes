## Text files and regular expressions

### Exercise 1 

import prideprejudice.txt and find if 'Chapter 2' is included in the text

```python
import csv
import os
import re # will help find chapters 
```

```python
pride_predList = []
with open("data/prideprejudice.txt", "r") as file:
    pride_pred_text = file.read() # read() reads the whole file and helps when searching whole file for something
    my_string = re.findall("\\bChapter 2\\b", pride_pred_text)

    if my_string:
        print("Chapter 2 is in the text")
    else:
        print("Chapter 2 is NOT in the text") 
                           

    # re.findall() - scans thru entire text and returns list of all matches
    # \b - word boundry
    # when using re.findall, must enclose thing ur finding in " "
```

```python
# can also use 
re.search("Chapter 2", pride_pred_text)
```

## Exercise 6
Write a function that recieves the name of the park and prints out it's state, establishment date, and description

```python
def np_info(np_name):
    file = open("data/nationalParks.json", 'r')
    data = json.loads(file.read())
    is_found = False
    for np in data:
        if np["title"] == np_name:
            is_found = True
            print("Park found.")
            print("The park is in: " + str(data[np]["state"]) + ", and was established in " + np["date_established"])
            print("Description: " + np["description"])
            return
        
    print("Park not found")   
    
```
