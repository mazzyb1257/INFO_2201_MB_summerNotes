### Exercise 3: 

import tv_shows.csv and find the 5 most recently made tv shows

```python
# 1 - import csv file and assign tv shows to items in a list
tvList = [] 
with open("data/tv_shows.csv", "r") as file:
    csvReader = csv.reader(file, delimiter = ',',quotechar = '"')
    next(csvReader) # skip header 
    
    # iterate thru file and add row to list
    for row in csvReader:
        tvList.append(row)
pprint.pprint(tvList)
```

```python
# 2 - extract shows and titles
# make empty dictionary for title, year pairs 
tv_yr_dict = {}

# iterate thru each show in tvList
for show in tvList:
    # extact the title and year 
    year_raw = show[2]
    title = show[1]

    # clean data
    year_clean = year_raw.replace("(","").replace(")","")
    year = int(year_clean[:4])
        # [:4] means give me items from start and up to but not including index 4 
    # assign the dictionary
    tv_yr_dict[show[1]] = year
print(pprint.pprint(tv_yr_dict))
```

```python
# 3 - get newest show value - 2015
newest_show = max(tv_yr_dict.values())
newest_show
```

```python
# 4 - we got the newest show year, so now we get the highest number 
top5 = sorted(tv_yr_dict, key = tv_yr_dict.get, reverse =True)[:5]
# .get sorts the dictionary by the values instead of the key
top5
```

```python
# 4 - not as simple way
topLyst = []
for key, value in tv_yr_dict.items():
    # .items returns all key-val pairs in the dictionary as tuples
    if value == newest_show: # newest_show is 2015
        topLyst.append(key)
# return only the values at index 0-5
topLyst[:5]
```

## Exercise 6
Write a function that recieves the name of the park and prints out it's state, establishment date, and description

```python
def np_info(np_name):
    import json
    with open("data/nationalParks.json", 'r') as f:
        data = json.loads(f.read())
    
    is_found = False
    
    for np in data:
        if np["title"] == np_name:
            is_found = True
            print("Park found.")
            
            # get state name 
            states = [state["title"] for state in np["states"]] # for state in np[states], states is the title 
            print("The park is in: " + ", ".join(states) + ", and was established in " + np["date_established_readable"])
            print("Description: " + np["description"])
            return
        
    print("Park not found")   
```
